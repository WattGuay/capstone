```python
#### Here is where I call in the data set

import pandas as pd

def lowercase_str(string):
    if isinstance(string, str):
        return string.lower()
    else:
        return string

# Read in the data set and apply the lowercase function to all columns
df = pd.read_csv('temp.dimval.csv')
df = df.applymap(lowercase_str)
```

```python
### here is where I cleaned the platform names 

# Define the mappings of values to be changed
plat_mappings = {
    'fb_ig': ['facebook/instagram', 'facebook & instagram', 'fb/ig','fb & ig'],
    'facebook': ['facebook', 'fb', 'facebok', 'facebook all', 'facebook ecommerce', 'facebook newsfeed', 'facebook video'], 
    'instagram': ['instagram all', 'instagram newsfeed', 'instagram video'], 
    'google': ['adwords', 'adwords', 'dv360', 'google','google ads', 'google adwords', 'google dv360'],
    'youtube': ['youtube', 'youtube', 'youtube (DV360)'],
    'connected_tv': ['connected tv','connected tv/fep'],
    'pinterest': ['pinterest', 'pinterest (beta)'],
    'trade desk': ['the trade desk', 'trade desk', 'tradedesk','ttd','ttd - iheart podcast'],
    'spotfiy': ['spotify', 'Spotify & iHeart', 'spotify, iheart, pandora', 'spotify/iheart'],
    'linkedin': ['linkedin', 'linkedin'],
    'bing': ['bing','bing ads'],
    'paid_search': ['paid search'],
    'paid_social': ['paid social'],
    'display': ['display', '- display', '--display', 'fy22 co-op round rock - display'],
    ### new additions 
    'native': ['native'],
    'twitter': ['twitter'],
    'performance': ['performance'],
    'standard': ['standard', 'standard rt'],
    'iheart': ['iHeart'],
    'dooh': ['dooh'],
    'tiktok': ['tiktok'],
    'snapchat': ['snapchat'],
    'ott': ['ott'],
    'pandora': ['pandora'],
    'olv': ['olv'],
    'sxm': ['sxm'],
    'twitch': ['twitch'],
    'nyt': ['nyt'],
    'tapjoy': ['tapjoy'],
    'hulu': ['hulu'],
    'quora': ['quora'],
    'tdd': ['tdd'],
    'fep': ['fep']
}



# Define a function to check if a string contains any of the specified values in any form
def contains_string(string, target_strings):
    return any(s in string.lower() for s in target_strings)

# Ask the user to input a column name
column_name = input("Enter the column name to scrape: ")

# Loop through the specified column and modify the values in-place
if df[column_name].dtype == 'object':  # Only modify object-type columns
    df[column_name] = df[column_name].apply(lambda x: next((k for k, v in plat_mappings.items() 
                                                            if contains_string(str(x), v)), x))

# Skip over non-object (e.g. float) values
else:
    pass

# Save the result to a new file
df.to_csv('p_cleaned.csv', index=False)
```
